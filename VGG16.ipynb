```python
import os
# 데이터 경로 설정
base_dir = "cars/cars_reduce"
train_dir = os.path.join(base_dir, 'Training')
val_dir = os.path.join(base_dir, 'Validation')

# 라벨 데이터를 저장할 리스트
tr_data = []
test_data = []
# 폴더 탐색
for car_type in os.listdir(train_dir):
  car_type_path = os.path.join(train_dir, car_type)

  # 차종 폴더인지 확인
  if os.path.isdir(car_type_path):
    for brand in os.listdir(car_type_path):
       brand_path = os.path.join(car_type_path, brand)

       # 브랜드 폴더인지 확인
       if os.path.isdir(brand_path):
        for image in os.listdir(brand_path):
          image_path = os.path.join(brand_path, image)

          # 모델 폴더인지 확인
          # if os.path.isfile(model_path):
          #   for image in os.listdir(model_path):
          #     image_path = os.path.join(model_path, image)

              # 이미지 파일인지 확인 (확장자 검사)
          if os.path.isfile(image_path) and image.lower().endswith(('.jpg', '.jpeg', '.png')):
                model = image_path.split('_')[1].split('-', -1)[0]
                  # 차종, 모델, 이미지 경로 추가
                tr_data.append({"CarType": car_type, "Brand" : brand, "Model": model, "ImagePath": image_path})

# 폴더 탐색
for car_type in os.listdir(val_dir):
  car_type_path = os.path.join(val_dir, car_type)

  # 차종 폴더인지 확인
  if os.path.isdir(car_type_path):
    for brand in os.listdir(car_type_path):
       brand_path = os.path.join(car_type_path, brand)

       # 브랜드 폴더인지 확인
       if os.path.isdir(brand_path):
        for image in os.listdir(brand_path):
          image_path = os.path.join(brand_path, image)

          # 모델 폴더인지 확인
          # if os.path.isfile(model_path):
          #   for image in os.listdir(model_path):
          #     image_path = os.path.join(model_path, image)

              # 이미지 파일인지 확인 (확장자 검사)
          if os.path.isfile(image_path) and image.lower().endswith(('.jpg', '.jpeg', '.png')):
                model = image_path.split('_')[1].split('-', -1)[0]
                  # 차종, 모델, 이미지 경로 추가
                test_data.append({"CarType": car_type, "Brand" : brand, "Model": model, "ImagePath": image_path})

# 데이터프레임 생성
tr_df = pd.DataFrame(tr_data)
test_df = pd.DataFrame(test_data)

# CSV 파일로 저장
output_file = "car_labels.csv"
tr_df.to_csv(output_file, index=False)
print(f"라벨링 데이터가 {output_file}에 저장되었습니다!")
```
